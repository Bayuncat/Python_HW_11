```python
from sympy import *
import numpy as np

x = symbols('x')
f = 0.6*x**3 + 5.5*x**2 + 10*x - 5

plot(f)
```

```python
x1, x2, x3 = solve(f)
```

```python
yprime = f.diff(x)
print(yprime)
```

```python
x4, x5 = solve(yprime)
```

```python
import math

y4 = 0.6*x4**3 + 5.5*x4**2 + 10*x4 - 5
y5 = 0.6*x5**3 + 5.5*x5**2 + 10*x5 - 5

print(f"Функция 'f = {f}' имеет корни:")
print(f"x1= {round(x1,2)}")
print(f"x2= {round(x2,2)}")
print(f"x3= {round(x3,2)}")
print(f"Функция f имеет вершины в точках {math.ceil(y4)} и {round(y5,2)} при значении х = {math.ceil(x4)} и x = {round(x5, 2)}")
print(f"Функция f возрастает на интервалах: от -∞ до {math.ceil(x4)} и от {round(x5, 2)} до +∞")
print(f"Функция f убывает на интервале от {math.ceil(x4)} до {round(x5, 2)}")
print(f"Функция f<0 при значениях x от -∞ до {round(x1,2)} и от {round(x2,2)} до {round(x3,2)}")
print(f"Функция f>0 при значениях x от {round(x1,2)} до {round(x2,2)} и от {round(x3,2)} до +∞")
```
